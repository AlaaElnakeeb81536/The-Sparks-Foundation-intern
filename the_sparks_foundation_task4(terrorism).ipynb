```python
#imprting library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
%matplotlib inline
```

```python
#showing data
terrorist_data=pd.read_csv('D:\globalterrorismdb_0718dist.csv',encoding='latin-1')
terrorist_data.head()
```

```python
#understanding data
terrorist_data.shape
```

```python
terrorist_data.columns
```

```python
terrorist_data.info(list)
```

```python
terrorist_data.describe()
```

```python
#selecting important col
terrorist_data=terrorist_data[['iyear','imonth','iday','country_txt','provstate','region_txt','attacktype1_txt','targtype1_txt','targsubtype1_txt','nkill','target1','latitude','longitude','gname','weaptype1_txt','nwound','city']]
```

```python
terrorist_data.shape
```

```python
#renaming label into understabded one
terrorist_data.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day',
                               'country_txt':'Country','region_txt':'Region',
                               'provstate':'State','city':'City',
                               'attacktype1_txt':'AttackType','targtype1_txt':'Targettype',
                               'targsubtype1_txt':'Targetsubtype',
                               'target1':'Target','nkill':'Killed',
                              'gname':'Group','weaptype1_txt':'Weapon_type','nwound':'Wounded'},inplace=True)
terrorist_data.columns
```

```python
#showing duplicated values
terrorist_data.duplicated().sum()
```

```python
#droping duplicatted values
terrorist_data.drop_duplicates(inplace=True)
terrorist_data.duplicated().sum()
```

```python
#checking for null values
terrorist_data.isnull().sum()
```

```python
#filling missing values
terrorist_data['Killed'].fillna(0,inplace=True)
```

```python
terrorist_data['Wounded'].fillna(0,inplace=True)
```

```python
terrorist_data['latitude'].fillna(0,inplace=True)
```

```python
terrorist_data['longitude'].fillna(0,inplace=True)
```

```python
terrorist_data.isnull().sum()
```

```python
#showing cleaned data
terrorist_data.head()
```

# Bivariete Analysis

```python
plt.subplot(1,2,1)
terrorist_data.groupby(['Region'])['Killed'].sum().plot(kind='bar')
plt.title('N.of Killed per Region',fontsize=21)
plt.ylabel('N.of Killed',fontsize=20)
plt.xlabel("Regions",fontsize=20)
plt.xticks(rotation=90)


plt.subplot(1,2,2)
terrorist_data.groupby(['Region'])['Wounded'].sum().plot(kind='bar')
plt.title('Number of Wounded per Region',fontsize=24)
plt.ylabel('N.of Wounded',fontsize=22)
plt.xlabel("Regions",fontsize=22)
plt.xticks(rotation=90)
plt.gcf().set_size_inches(17,10)
plt.tight_layout()
```

we can conclude that 'Middle East and North Africa' has the highest killed and wounded people by terrorrism

```python
plt.subplot(1,2,1)
terrorist_data.groupby(['Country'])['Killed'].sum().sort_values(ascending=False)[:10].plot(kind='bar')
plt.title('N.of Killed per Country',fontsize=21)
plt.ylabel('N.of Killed',fontsize=20)
plt.xlabel("Country",fontsize=20)
plt.xticks(rotation=90)


plt.subplot(1,2,2)
terrorist_data.groupby(['Country'])['Wounded'].sum().sort_values(ascending=False)[:10].plot(kind='bar')
plt.title('Number of Wounded per Country',fontsize=24)
plt.ylabel('N.of Wounded',fontsize=22)
plt.xlabel("Country",fontsize=22)
plt.xticks(rotation=90)
plt.gcf().set_size_inches(15,8)
plt.tight_layout()
```

we can conclude that Iraq is the country that has the most killed (between 70000 and 80000)and wounded people(>130000)

```python
plt.figure(figsize=(10,10))
sns.barplot(terrorist_data['Group'].value_counts()[1:11].values,terrorist_data['Group'].value_counts()[1:11].index)
plt.title('Top Terrorist Organizations ',fontsize=13)
plt.xlabel('N.of Attacks')
plt.ylabel('Name of Terrorism organization')
plt.xticks(rotation=90)
plt.show()
```
